```python
# 필요한 컬럼만 추출해서 df2를 준비
df2_reduced = df2[['ID', '기준년월', 'Segment']]

# ID와 기준년월 기준으로 병합 (left join)
df_merged = pd.merge(df1, df2_reduced, on=['ID', '기준년월'], how='left')
```

```python
# 병합한 원본데이터(df_merged)는 유지하고, 복사본(df_converted)을 생성
df_converted = df_merged.copy()

# '캠페인접촉건수_R12M' 범주를 숫자로 바꿔준다.
mapping = {
    '1회 이상': 1,
    '5회 이상': 2,
    '10회 이상': 3,
    '15회 이상': 4,
    '20회 이상': 5,
    '25회 이상': 6
}

df_converted['캠페인접촉건수_R12M'] = df_converted['캠페인접촉건수_R12M'].replace(mapping)
df_converted['캠페인접촉건수_R12M'].value_counts()
```

```python
# '캠페인접촉일수_R12M'의 고유값 분포(범주 구성 및 빈도) 확인
df_merged['캠페인접촉일수_R12M'].value_counts()
```

```python
# 복사본(df_converted)데이터를 활용한다.
# '캠페인접촉일수_R12M' 범주를 숫자로 바꿔준다.
mapping = {
    '1일 이상': 1,
    '5일 이상': 2,
    '10일 이상': 3,
    '15일 이상': 4,
    '20일 이상': 5
}

df_converted['캠페인접촉일수_R12M'] = df_converted['캠페인접촉일수_R12M'].replace(mapping)
df_converted['캠페인접촉일수_R12M'].value_counts()
```

```python
# 'Segment'의 고유값 분포(범주 구성 및 빈도) 확인
df_merged['Segment'].value_counts()
```

```python
# 복사본(df_converted)데이터를 활용한다.
# Segment 데이터를 숫자데이터로 바꾼다.
mapping = {
    'A': 1,
    'B': 2,
    'C': 3,
    'D': 4,
    'E': 5
}

df_converted['Segment'] = df_converted['Segment'].replace(mapping)
df_converted['Segment'].value_counts()
```

```python
# 범주형데이터가 숫자데이터로 바뀌었는지 확인
df_converted.select_dtypes(include=['object', 'category']).columns
```

```python
# 요약된 결과가 아닌, 모든 결과를 보기 위한 코드
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
# ID 열 제외하고 상관계수 계산
df_corr = df_converted.drop(columns=['ID']).corr()

# 특정 변수('Segment')와의 상관계수만 추출
segment_corr = df_corr['Segment'].drop('Segment')  # 자기 자신 제외

# 절댓값 기준 내림차순 정렬
sorted_corr = segment_corr.loc[segment_corr.abs().sort_values(ascending=False).index]

# 지수표기법이 아닌 상관계수값으로 보기 위한 코드, 전체에 적용되는게 아닌 특정 결과만 소수로 변환해서 보고 싶을 때 사용
display(sorted_corr.apply(lambda x: f"{x:.4f}"))
```

```python
# NaN인 상관계수만 따로 확인
nan_corr = sorted_corr[sorted_corr.isna()]
display(nan_corr)
```

```python
# 결과로 Nan이 나온 변수의 고유값 개수 확인
for col in nan_corr.index:
    unique_count = df_converted[col].nunique()
    print(f"'{col}' -> 고유값 개수: {unique_count}")
```
